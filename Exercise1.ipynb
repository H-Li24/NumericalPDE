$$
u' = f(t,u)  \quad 0<t \leq 1, \quad u(0)=1
$$

with right hand side
$$
f(t,u) = -u + t^3 + 3 t^2
$$

The exact solution
$$
u(t) = e^{-t} + x^3
$$

Set up

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the right side and the exact solution

```python
def f(t,u):
    return -u + t ** 3 + 3 * t ** 2
```

```python
def u_exact(t):
    return np.exp(-t) + t ** 3
```

step 1: generate mesh

```python
NT = 10 # number of interval
NP = NT + 1 # number of points
a_left = 0
b_right = 1
h = (b_right-a_left)/NT # step sie
print(h)
t = np.zeros([NP,1], dtype = float)
for n in range(NP): #from 0 to NT, NT+1=NP points
    t[n] = n * h
print(t)
```

step 2: Euler method
$$
u_{n+1} = u_n + hf(t_n, u_n), \quad n = 0, 1, ..., NP -1
$$


```python
uh = np.zeros([NP,1], dtype = float)
uh[0] = 1.0
for n in range(NT): #from 1 to NP, NT-1+1=NT=NP-1 points
    uh[n+1] = uh[n] + h * f(t[n], uh[n])
print(uh)
```

step 3: Compute the exact solution

```python
u = np.zeros([NP,1], dtype = float)
for n in range(NP):
    u[n] = u_exact(t[n])
print(u)
```

```python
plt.plot(t, uh)
plt.plot(t, u)
plt.show()
```

Step 4: Compute the error

```python
err = np.abs(u - uh)
plt.plot(t, err)
plt.show()
```
